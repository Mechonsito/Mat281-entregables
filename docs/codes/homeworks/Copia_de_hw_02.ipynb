```python
import pandas as pd
import os
import zipfile

# Define la ruta a la carpeta de datos en Google Drive
data_path = '/content/drive/MyDrive/Tarea 2 aplica/'
zip_file_path = os.path.join(data_path, 'learnplatform-covid19-impact-on-digital-learning.zip')
unzip_path = '/tmp/learnplatform_data/'
# Directorio temporal para descomprimir archivos


# Crear el directorio temporal si no existe
os.makedirs(unzip_path, exist_ok=True)

# Descomprimir los datos
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(unzip_path)

# Definir las rutas a los datos descomprimidos
engagement_data_path = os.path.join(unzip_path, 'engagement_data')
districts_info_path = os.path.join(unzip_path, 'districts_info.csv')
products_info_path = os.path.join(unzip_path, 'products_info.csv')


# Cargar districts_info.csv y products_info.csv
districts_df = pd.read_csv(districts_info_path)
products_df = pd.read_csv(products_info_path)

# Cargar datos de participación de archivos CSV en la carpeta engagement_data
engagement_dfs = []

# Listar archivos directamente en engagement_data_path
engagement_files = os.listdir(engagement_data_path)

for csv_file in engagement_files:
    # Verificar si el archivo es un CSV y no un archivo oculto
    if csv_file.endswith('.csv') and not csv_file.startswith('.'):
        file_path = os.path.join(engagement_data_path, csv_file)
        try:
            df = pd.read_csv(file_path)
            # Extraer district_id del nombre del archivo (ej., '3668.csv' -> 3668)
            df['district_id'] = int(os.path.splitext(csv_file)[0])
            engagement_dfs.append(df)
        except Exception as e:
            print(f"Error reading file {file_path}: {e}")


# Concatenar todos los dataframes de participación
if engagement_dfs:
    engagement_df = pd.concat(engagement_dfs, ignore_index=True)
    print("Engagement data loaded and concatenated successfully.")
else:
    engagement_df = pd.DataFrame()
    # Crear un DataFrame vacío si no se cargaron archivos
    print("No engagement data files found.")

# Mostrar las primeras filas de cada dataframe para confirmar la carga
print("Districts Info:")
display(districts_df.head())

print("\nProducts Info:")
display(products_df.head())

print("\nEngagement Data:")
display(engagement_df.head())
```

### Integración de Datos

```python
# Cambiar el nombre de 'LP ID' en products_df a 'lp_id' para que coincida con engagement_df si es necesario (según la descripción de los datos)
# Verificar los nombres de las columnas en engagement_df primero para confirmar el nombre de la columna
if 'lp_id' in engagement_df.columns and 'LP ID' in products_df.columns:
    products_df.rename(columns={'LP ID': 'lp_id'}, inplace=True)
elif 'LP ID' in engagement_df.columns and 'LP ID' in products_df.columns:
     # Si la columna de engagement_df también es 'LP ID', no se necesita cambiar el nombre, pero es bueno ser explícito
     pass
elif 'lp_id' not in engagement_df.columns and 'LP ID' in products_df.columns:
     print("Warning: 'lp_id' column not found in engagement_df. Cannot merge with products_df.")
     # Advertencia: columna 'lp_id' no encontrada en engagement_df. No se puede fusionar con products_df.


# Fusionar engagement_df con districts_df en 'district_id'
# Suponiendo que 'district_id' es la clave correcta y existe en ambos dataframes
if 'district_id' in engagement_df.columns and 'district_id' in districts_df.columns:
    merged_df = pd.merge(engagement_df, districts_df, on='district_id', how='left')
    print("Merged engagement_df with districts_df.")
    # engagement_df fusionado con districts_df.
else:
    merged_df = engagement_df.copy()
    print("Could not merge engagement_df with districts_df: 'district_id' column not found in one or both dataframes.")
    # No se pudo fusionar engagement_df con districts_df: la columna 'district_id' no se encuentra en uno o ambos dataframes.


# Fusionar el resultado con products_df en 'lp_id' (o 'LP ID' dependiendo de los nombres de columna reales)
# Suponiendo que 'lp_id' (o 'LP ID') es la clave correcta y existe en ambos dataframes después de un posible cambio de nombre
lp_id_col_products = 'lp_id' if 'lp_id' in products_df.columns else ('LP ID' if 'LP ID' in products_df.columns else None)
lp_id_col_engagement = 'lp_id' if 'lp_id' in merged_df.columns else ('LP ID' if 'LP ID' in merged_df.columns else None)


if lp_id_col_products and lp_id_col_engagement:
     if lp_id_col_products != lp_id_col_engagement:
         # Asegurar que los nombres de las columnas coincidan antes de fusionar
         merged_df.rename(columns={lp_id_col_engagement: 'lp_id'}, inplace=True)
         products_df.rename(columns={lp_id_col_products: 'lp_id'}, inplace=True)
         lp_id_col = 'lp_id'
     else:
         lp_id_col = lp_id_col_products

     merged_df = pd.merge(merged_df, products_df, on=lp_id_col, how='left')
     print("Merged with products_df.")
     # Fusionado con products_df.
elif not lp_id_col_products:
    print("Could not merge with products_df: 'lp_id' or 'LP ID' column not found in products_df.")
    # No se pudo fusionar con products_df: la columna 'lp_id' o 'LP ID' no se encuentra en products_df.
elif not lp_id_col_engagement:
     print("Could not merge with products_df: 'lp_id' or 'LP ID' column not found in merged_df (from engagement_df).")
     # No se pudo fusionar con products_df: la columna 'lp_id' o 'LP ID' no se encuentra en merged_df (desde engagement_df).


# Mostrar las primeras filas del dataframe fusionado
print("\nMerged Data:")
display(merged_df.head())

# Mostrar información y recuentos de valores nulos para el dataframe fusionado
print("\nExploring merged_df:")
merged_df.info()
display(merged_df.isnull().sum())
```

### Limpieza y Preprocesamiento de Datos

```python
# Convertir la columna 'time' a objetos datetime
merged_df['time'] = pd.to_datetime(merged_df['time'])

# Manejar valores faltantes
# Para columnas numéricas, rellenar con la mediana o la media (dependiendo de la distribución)
# Para columnas categóricas, rellenar con la moda o un marcador de posición como 'Unknown'

# Verifiquemos los valores nulos nuevamente para decidir la estrategia
print("Missing values before handling:")
display(merged_df.isnull().sum())

# Rellenar los valores numéricos faltantes (engagement_index, pct_access - aunque pct_access tiene muy pocos)
# Usando la mediana para engagement_index ya que podría estar sesgado
median_engagement = merged_df['engagement_index'].median()
merged_df['engagement_index'].fillna(median_engagement, inplace=True)

# pct_access tiene muy pocos valores faltantes, se podría rellenar con 0 o la mediana. Usemos la mediana por ahora.
median_pct_access = merged_df['pct_access'].median()
merged_df['pct_access'].fillna(median_pct_access, inplace=True)


# Rellenar los valores categóricos faltantes con un marcador de posición 'Unknown' o la moda
# 'state', 'locale', 'pct_black/hispanic', 'pct_free/reduced', 'county_connections_ratio', 'pp_total_raw'
categorical_cols = ['state', 'locale', 'pct_black/hispanic', 'pct_free/reduced', 'county_connections_ratio', 'pp_total_raw',
                    'URL', 'Product Name', 'Provider/Company Name', 'Sector(s)', 'Primary Essential Function']

for col in categorical_cols:
    if col in merged_df.columns:
        merged_df[col].fillna('Unknown', inplace=True)

# Verificar que los valores faltantes se hayan manejado
print("\nMissing values after handling:")
display(merged_df.isnull().sum())

print("\nDataFrame info after cleaning:")
merged_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular el índice de participación promedio por día
daily_engagement = merged_df.groupby('time')['engagement_index'].mean().reset_index()

# Visualizar la participación digital promedio a lo largo del tiempo
plt.figure(figsize=(15, 7))
sns.lineplot(data=daily_engagement, x='time', y='engagement_index')
plt.title('Average Digital Engagement Over Time (2020)')
plt.xlabel('Date')
plt.ylabel('Average Engagement Index')
plt.grid(True)
plt.show()
```

**Análisis de la participación digital promedio a lo largo del tiempo:**

El gráfico muestra la evolución del índice de participación digital promedio a lo largo del año 2020. Podemos observar picos y valles que podrían correlacionarse con eventos escolares, vacaciones, o el impacto directo de la pandemia y las medidas de confinamiento. Una caída notable podría indicar períodos de menor actividad en las plataformas digitales, mientras que los picos podrían sugerir una mayor dependencia del aprendizaje en línea. Es crucial investigar los eventos clave de 2020 (cierres de escuelas, reaperturas, fases de la pandemia) para interpretar mejor estas fluctuaciones.

```python
# Calcular el índice de participación promedio para cada función esencial primaria
engagement_by_product_type = merged_df.groupby('Primary Essential Function')['engagement_index'].mean().reset_index()

# Ordenar por participación promedio para una mejor visualización
engagement_by_product_type = engagement_by_product_type.sort_values('engagement_index', ascending=False)

# Visualizar la participación promedio por tipo de producto
plt.figure(figsize=(12, 8))
sns.barplot(data=engagement_by_product_type, x='engagement_index', y='Primary Essential Function', palette='viridis')
plt.title('Average Digital Engagement by Primary Essential Function')
plt.xlabel('Average Engagement Index')
plt.ylabel('Primary Essential Function')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Rellenar los valores faltantes de 'Age' con la mediana de edad del conjunto de entrenamiento.
median_age = train_df['Age'].median()
train_df['Age'].fillna(median_age, inplace=True)
test_df['Age'].fillna(median_age, inplace=True)

# Rellenar los valores faltantes de 'Embarked' en train_df con el puerto de embarque más frecuente.
most_frequent_embarked = train_df['Embarked'].mode()[0]
train_df['Embarked'].fillna(most_frequent_embarked, inplace=True)

# Rellenar los valores faltantes de 'Fare' en test_df con la mediana de la tarifa del conjunto de entrenamiento.
median_fare = train_df['Fare'].median()
test_df['Fare'].fillna(median_fare, inplace=True)

# Crear una nueva característica 'FamilySize'.
train_df['FamilySize'] = train_df['SibSp'] + train_df['Parch'] + 1
test_df['FamilySize'] = test_df['SibSp'] + test_df['Parch'] + 1

# Crear una nueva característica 'IsAlone'.
train_df['IsAlone'] = (train_df['FamilySize'] == 1).astype(int)
test_df['IsAlone'] = (test_df['FamilySize'] == 1).astype(int)

# Eliminar columnas irrelevantes.
train_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp', 'Parch'], axis=1, inplace=True)
test_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp', 'Parch'], axis=1, inplace=True)


# Convertir 'Sex' en representación numérica.
train_df['Sex'] = train_df['Sex'].map({'male': 0, 'female': 1})
test_df['Sex'] = test_df['Sex'].map({'male': 0, 'female': 1})

# Convertir 'Embarked' en representación numérica usando one-hot encoding.
train_df = pd.get_dummies(train_df, columns=['Embarked'], prefix='Embarked', drop_first=True)
test_df = pd.get_dummies(test_df, columns=['Embarked'], prefix='Embarked', drop_first=True)


# Mostrar las primeras filas y la información del train_df preprocesado.
display(train_df.head())
train_df.info()
```

**Preprocesamiento de Datos del Titanic:**

En este paso, hemos abordado los valores faltantes rellenando 'Age' con la mediana, 'Embarked' con la moda, y 'Fare' con la mediana. Esto nos permite trabajar con conjuntos de datos completos. También hemos creado nuevas características: 'FamilySize' (suma de hermanos/cónyuges y padres/hijos más uno) y 'IsAlone' (indicando si el pasajero viajaba solo). Estas características compuestas a menudo capturan información útil para la predicción. Se eliminaron columnas menos relevantes como 'PassengerId', 'Name', 'Ticket' y 'Cabin'. Finalmente, las variables categóricas 'Sex' y 'Embarked' se convirtieron a formato numérico utilizando mapeo binario para 'Sex' y one-hot encoding para 'Embarked'. El `info()` del dataframe preprocesado confirma que no hay valores nulos y que las columnas están en formatos adecuados para el modelado.

```python
from sklearn.linear_model import LogisticRegression

X_train = train_df.drop('Survived', axis=1)
y_train = train_df['Survived']

model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred_train = model.predict(X_train)

accuracy = accuracy_score(y_train, y_pred_train)
precision = precision_score(y_train, y_pred_train)
recall = recall_score(y_train, y_pred_train)
f1 = f1_score(y_train, y_pred_train)

print(f'Training Accuracy: {accuracy:.4f}')
print(f'Training Precision: {precision:.4f}')
print(f'Training Recall: {recall:.4f}')
print(f'Training F1-score: {f1:.4f}')
```

**Modelado y Evaluación (Regresión Logística):**

Hemos entrenado un modelo de Regresión Logística simple utilizando las características preprocesadas. La evaluación en el conjunto de entrenamiento muestra una precisión de aproximadamente 80.58%. Las métricas de precisión, recall y F1-score proporcionan una visión más detallada del rendimiento del modelo. Una precisión del 77.70% significa que, de las predicciones positivas (sobrevivió), el 77.70% fueron correctas. Un recall del 69.30% indica que el modelo identificó correctamente al 69.30% de los pasajeros que realmente sobrevivieron. El F1-score (73.26%) es una media armónica de precisión y recall. Estos resultados iniciales son razonables para un modelo simple, pero podrían mejorarse con ingeniería de características más avanzada o el uso de modelos más complejos.
